## Visualisation d’un nuage de points et exploration des variables 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
#print (os.path.abspath(os.path.dirname('house_pricing.csv')))
house = pd.read_csv(os.path.abspath(os.path.dirname('house_pricing.csv'))+"/house_pricing.csv",",")
```

```python
house.head()
```

```python
house.describe()
```

```python
house.info()
```

```python
#plt.figure(figsize=(7,7))
n1 = plt.plot(house.LotArea,house.SalePrice,'.')
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
```

```python
#plt.figure(figsize=(7,7))
houseReduit = house[(house.LotArea<=20000)&(house.SalePrice<=500000)]
n2 = plt.plot(houseReduit.LotArea,houseReduit.SalePrice,'.')
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
```

```python
n3 = plt.plot(house.LotFrontage,house.LotArea,'.')
plt.xlabel("LotFrontage")
plt.ylabel("LotArea")
```

```python
houseReduit1 = house[(house.LotFrontage<=200)&(house.LotArea<=100000)]
n4 = plt.plot(houseReduit1.LotFrontage,houseReduit1.LotArea,'.')
plt.xlabel("LotFrontage")
plt.ylabel("LotArea")
```

## Visualisation de relations continues 

```python
sales = pd.read_csv(os.path.abspath(os.path.dirname('sales_predictions.csv'))+"/sales_predictions.csv",",")
sales.head()
```

```python
n1 = sns.relplot(x='date', y='item_price', data=sales)
```

```python
salesReduced = sales.sample(n=50)
salesReduced.head()
salesReduced.info()
```

```python
plt.figure(figsize=(20,6))
n1 = sns.relplot(x='date', y='item_price', data=salesReduced,sizes=(20,6))
plt.show()
```

On voit que les dates sont mélangées entre elles et on arrive pas a bien distinguer les données par rapport à la date.

```python
salesReduced.date =(pd.to_datetime(salesReduced.date))
```

```python
salesReduced.head()
```

```python
plt.figure(figsize=(20,6))
n1 = sns.relplot(x='date', y='item_price', data=salesReduced,sizes=(20,6))
plt.show()
```
